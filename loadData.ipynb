## 3. Nettoyage et Analyse Exploratoire des Données (EDA)

Cette section automatise le processus de nettoyage et d'analyse descriptive pour l'ensemble des fichiers chargés. L'objectif est de garantir l'intégrité des données et d'optimiser la structure des DataFrames avant l'analyse approfondie.

### Méthodologie appliquée par fichier :

1. **Définition de l'unité d'observation (*Row definition*)** :
   - Identification de la granularité technique d'une ligne (ex: une observation unique pays/année/indicateur).
2. **Évaluation de la volumétrie (*Shape*)** :
   - Calcul du nombre de lignes (*records*) et de colonnes (*features*) pour quantifier le dataset.
3. **Traitement des Redondances (*Deduplication*)** :
   - Détection et suppression des doublons pour éviter de biaiser les futurs calculs statistiques (moyennes, sommes).
4. **Analyse de la complétude (*Missing Values*)** :
   - Calcul de la proportion de valeurs manquantes (`NaN`) par colonne pour évaluer la fiabilité de chaque variable.
5. **Optimisation du Dataset (*Pruning*)** :
   - Suppression des colonnes jugées inutilisables (ex: colonnes techniques vides ou colonnes ayant plus de 90% de valeurs manquantes).
      - #### Justification technique :

        * **Significativité statistique** : Une variable renseignée à moins de 10% ne permet pas d'extraire des tendances représentatives et introduit un "bruit" analytique (*statistical noise*) qui fausse les mesures de tendance centrale comme les moyennes et les écart-types.
        * **Pertinence temporelle** : Dans le fichier `EdStatsData.csv`, ce seuil permet d'éliminer les projections à très long terme (ex: 2070-2100) qui sont quasi-intégralement vides, tout en conservant les données historiques réelles indispensables à l'analyse.
        * **Performance (*Memory Management*)** : La suppression de ces colonnes réduit l'empreinte mémoire du DataFrame, ce qui accélère les calculs ultérieurs (calculs vectorisés), une pratique essentielle sur des jeux de données dépassant les 800 000 lignes.

6. **Analyse Descriptive (*Numerical & Categorical Features*)** :
   - **Colonnes Numériques** : Application de `.describe()` pour obtenir les mesures de tendance centrale et de dispersion (Min, Max, Moyenne, Quartiles).
   - **Colonnes Catégorielles** : Calcul des occurrences via `.value_counts()` pour identifier les modalités dominantes et les déséquilibres potentiels.

> **Note technique :** Afin de garantir un rendu visuel stable dans l'IDE et un export PDF professionnel, les résultats catégoriels sont convertis en structures tabulaires (*DataFrames*) avant d'être affichés via la fonction `display()`.

> **Traitement des redondances**
> * **.duplicated()** : Cette méthode parcourt le DataFrame et renvoie un masque booléen (une série de `True` ou `False`). Elle marque True pour chaque ligne dont toutes les valeurs sont strictement identiques à une ligne précédente.
> * **.sum()** : En Python, les valeurs booléennes sont traitées numériquement (`True = 1` et `False = 0`). En appliquant `.sum()` sur le résultat de duplicated(), nous obtenons instantanément le nombre total de lignes dupliquées.

```python
duplicate_count = dataframes[0]['data'].duplicated().sum()
display(Markdown(f"### 3.3. Le fichier : {dataframes[0]['name']} posséde {duplicate_count} lignes dupliquées"))
```

> * **.drop_duplicates()** : Si le compte est supérieur à zéro, cette méthode permet de supprimer les copies pour ne conserver qu'une seule occurrence unique de chaque ligne.

```python
if duplicate_count > 0:
    dataframes[0]['data'] = dataframes[0]['data'].drop_duplicates()
```

### 3.4. Calcul du pourcentage de valeurs manquantes par colonnes et affichage dans un nouveau dataframe trié par pourcentage décroissant.

> **Analyse de la complétude** :
> * **.isnull() ou .isna()** :  Cette méthode génère un masque booléen où chaque cellule vide (NaN) est marquée comme True.
> * **len()** : En divisant la somme des valeurs nulles par la longueur totale du DataFrame (`len(df)`), nous obtenons une proportion que nous multiplions par 100 pour obtenir un pourcentage lisible.
> * **pd.DataFrame()** : Nous reconstruisons un nouveau tableau pour synthétiser ces résultats, ce qui facilite la lecture par rapport à une simple liste brute.
> * **.sort_values()** : Cette méthode permet de trier le résultat. En utilisant ascending=False, nous plaçons les colonnes les plus vides en haut de tableau pour identifier immédiatement les variables inutilisables.

```python
percent_missing = dataframes[0]['data'].isnull().sum() * 100 / len(dataframes[0]['data'])
missing_value = pd.DataFrame({'column_name' : dataframes[0]['data'].columns, 'percent_missing' : percent_missing}).sort_values(by='percent_missing', ascending=False)
display(missing_value)
```

### 3.5. Néttoyage des collones atteignant plus de 90% de valeurs manquantes

> **Optimisation du Data Set**
> * **.dropna() avec un seuil** : axis=1 : Ce paramètre indique à Pandas d'agir sur les colonnes (l'axe vertical) plutôt que sur les lignes.
> * **thresh=limit** : C'est le paramètre de seuil (threshold). Il définit le nombre minimum de valeurs non-nulles requises pour que la colonne soit conservée.
> * **Logique de calcul** : Si on définis limit à 10% de la longueur totale (len(df) * 0.1), toute colonne ayant plus de 90% de données manquantes sera supprimée.
> * **Utilité** : Cette méthode est plus robuste que de supprimer les colonnes par leur nom, car elle s'adapte dynamiquement au contenu réel de chaque fichier sans hardcoding.

```python
limit = len(dataframes[0]['data']) * 0.1
dataframes[0]['data'] = dataframes[0]['data'].dropna(axis=1, thresh=limit)
display(dataframes[0]['data'].head())
```

```python
display(Markdown(f"### 3.6.1. Obtention des mesures de tendances centrales et de dispersions : {dataframes[0]['name']}"))
```

> **Analyse descriptive (Numerical feature)** :
> * **select_dtypes(include=['number'])** : Cette méthode filtre le DataFrame pour ne conserver que les colonnes de type numérique (entiers ou flottants). C'est une sécurité indispensable avant de lancer des calculs mathématiques qui échoueraient sur du texte.
> * **.empty** : Un attribut qui renvoie True si le DataFrame résultant ne contient aucune donnée. Nous l'utilisons pour conditionner l'affichage et éviter des erreurs de rendu.
> * **.describe()** : La fonction maîtresse de l'analyse descriptive. Elle génère automatiquement un tableau de synthèse incluant :
> * count : Le nombre de valeurs renseignées (non-nulles).
> * mean : La moyenne arithmétique.
> * std : L'écart-type, mesurant la dispersion des données autour de la moyenne.
> * min / max : Les valeurs extrêmes du jeu de données.
> * 25%, 50% (médiane), 75% : Les quartiles, permettant de comprendre la répartition des données.

```python
numeric_df = dataframes[0]['data'].select_dtypes(include=['number'])

if not numeric_df.empty:
    display(dataframes[0]['data'].describe(include=[np.number]))
else :
    display(Markdown(f" Le fichier : {dataframes[0]['name']} ne posséde pas de valeur numérique"))
```

```python
df_source = dataframes[0]['data']
cat_cols = df_source.select_dtypes(include=['object', 'string']).columns

report_chunks = []

for col in cat_cols:
    counts = df_source[col].value_counts().head(5).to_frame()

    counts = counts.reset_index()
    counts.columns = ['Valeur', 'Nombre']
    counts.insert(0, 'Variable', col)
    report_chunks.append(counts)

final_report = pd.concat(report_chunks, ignore_index=True)

display(Markdown(f"### 3.6.2. Rapport global des occurrences : {dataframes[0]['name']}"))
```

> **Analyse des variables catégorielles**
> * **select_dtypes(include=['object', 'string'])** : Cette méthode cible exclusivement les colonnes textuelles (catégorielles).
> * **value_count().head(5)** : Pour chaque colonne, nous calculons la fréquence d'apparition de chaque texte et ne conservons que le "Top 5".
> * **reset_index et insert()** : Ces manipulations permettent de transformer un résultat de calcul (Series) en un tableau propre, en réintégrant le nom de la variable d'origine comme une colonne à part entière.
> * **pd.concat()** : Cette fonction est l'équivalent d'un "collage" de tableaux. Elle rassemble tous les petits rapports individuels (report_chunks) en un seul grand DataFrame final pour une lecture centralisée.
> * **pd.option_context** : C'est un "gestionnaire de contexte" (utilisé avec le mot-clé with). Il permet de modifier temporairement les réglages de Pandas uniquement pour le bloc de code qui suit.
> * **'display.max_rows'** : En réglant cette option sur None, on indique à Pandas qu'il n'y a plus de limite au nombre de lignes à afficher.
> * **Avantage** : Une fois sorti du bloc with, Pandas reprend ses réglages par défaut, ce qui évite d'encombrer le Notebook pour les autres affichages plus petits.

```python
with pd.option_context('display.max_rows', None):
    display(final_report)
```

### 3.6. Note sur l'analyse descriptive de EdStatsData.csv

Bien que la méthode `.describe()` s'exécute sans erreur technique sur ce fichier, les résultats statistiques globaux (moyenne, écart-type) sont **analytiquement inutilisables** en l'état pour les raisons suivantes :

* **Hétérogénéité des indicateurs (*Mixed Scales*)** : Chaque colonne "Année" mélange des données de natures totalement différentes. Faire la moyenne entre un PIB (en milliers de milliards), une population (en milliards) et un taux d'alphabétisation (en pourcentage) génère un chiffre dépourvu de sens métier.
* **Biais de dispersion (*Variance Bias*)** : L'écart-type (*standard deviation*) extrêmement élevé observé dans les résultats (ex: $1.2 \times 10^{11}$ pour 1970) confirme que les données ne suivent pas une distribution commune. Ce "bruit" statistique masque la réalité de chaque indicateur individuel.
* **Interprétation** : Pour obtenir des statistiques descriptives cohérentes, il est impératif d'effectuer un **filtrage préalable** sur la colonne `Indicator Code` afin d'isoler une seule métrique avant d'appliquer des calculs d'agrégation.

> **Conclusion de l'audit** : Ce tableau global n'est conservé ici qu'à titre de validation technique de la lecture des données numériques. Ce jeu de données servira exclusivement de base à une analyse par filtrage sélectif (méthodes `.loc` ou `.query`). Cette approche est la seule permettant de garantir la pertinence des calculs en isolant chaque métrique de son contexte d'origine.

```python
df_source = dataframes[1]['data']
cat_cols = df_source.select_dtypes(include=['object', 'string']).columns

report_chunks = []

for col in cat_cols:
    counts = df_source[col].value_counts().head(5).to_frame()

    counts = counts.reset_index()
    counts.columns = ['Valeur', 'Nombre']
    counts.insert(0, 'Variable', col)
    report_chunks.append(counts)

final_report = pd.concat(report_chunks, ignore_index=True)

display(Markdown(f"### 3.6.2. Rapport global des occurrences : {dataframes[1]['name']}"))
```

## 6.1. Développement d'un indicateur de densité :

Pour évaluer la fiabilité de notre dataset, nous avons développé une fonction utilitaire permettant de mesurer la densité des données. Cette approche évite la répétition de code (principe DRY - Don't Repeat Yourself) et garantit une mesure homogène sur tous les axes du DataFrame.

**Justification de la logique appliquée :**
   - `.notnull()` : Cette méthode crée un masque booléen identifiant chaque cellule contenant une information réelle. Les valeurs NaN (manquantes) sont ignorées.
   - `.mean(axis=axis)` : En calculant la moyenne de ce masque booléen (True valant 1 et False valant 0), nous obtenons directement le taux de remplissage exprimé entre 0 et 1.
   - Flexibilité de l'axe (`axis`) :
      - `axis=0` (par colonne) : Permet d'identifier les années les plus riches en données pour l'ensemble des pays
      - `axis=1` (par ligne) : Permet de calculer un score de fiabilité pour chaque indicateur spécifique, en mesurant sa régularité sur la plage temporelle choisie.

> **Note métier** : Cette fonction nous permet de ne retenir que les segments ayant une masse critique suffisante pour que les conclusions stratégiques d'Academy soient statistiquement significatives.

```python
def calculate_data_fill_rate(df_analys, axis=0):
    return df_analys.notnull().mean(axis=axis)
```

1. **Fonctionnement technique : Le Slicing avec `.loc`**
    - `.loc` : C'est une méthode d'accès par étiquettes (labels).
    - `:` : Le premier argument (avant la virgule) signifie qu'on sélectionne toutes les lignes.
    - `'2010':'2015'` : Le deuxième argument définit la plage de colonnes. Grâce à l'opérateur `:`, Pandas récupère toutes les colonnes situées entre 2010 et 2015 inclus.

```python
df_years_only = dataframes[2]['data'].loc[:, '2010':'2015']
display(df_years_only)
```

> **Fonctionnement technique** :
> - `calculate_data_fill_rate(..., axis=0)` : En passant l'argument axis=0, on indiques à la fonction de calculer la moyenne verticalement (colonne par colonne).
> - `.to_frame(name='...')` : Par défaut, le résultat est une Series Pandas (une simple liste indexée). Cette méthode la convertit en un DataFrame, ce qui facilite la lecture et l'exportation future du rapport.

```python
data_fill_rate_per_year = calculate_data_fill_rate(df_years_only, axis=0)
display(data_fill_rate_per_year.to_frame(name='Ratio de complétude'))
```

## 6.2. Stratégie de sélection finale

> **Fonctionnement technique** :
>
> Le groupby est une opération en trois étapes souvent appelée **Split-Apply-Combine**.
> - **Split** : Pandas rassemble toutes les lignes qui ont le même Indicator Name (par exemple, toutes les lignes "Internet users" de tous les pays).
> - **Apply** : La fonction `.count()` est appliquée à chaque groupe.
>   - Note importante : `.count()` ne compte que les valeurs non-nulles. Si un pays n'a pas de donnée pour 2010, il n'est pas compté.
> - **Combine** : Pandas rassemble les résultats dans un nouveau tableau où chaque ligne est un indicateur unique, et chaque colonne contient le nombre de pays l'ayant renseigné.
>
> **Mesure de la couverture géographique** : On ne regarde plus si une cellule est vide, on regarde si l'indicateur est global. Un indicateur présent dans 200 pays a beaucoup plus de valeur qu'un indicateur présent dans seulement 5 pays.
>
> **Tri stratégique** : Le `sort_values` permet d'isoler immédiatement les "indicateurs stars" (comme Internet users ou Secondary enrolment) qui seront les piliers de la stratégie d'Academy.

# 8. Analyse des corrélations et réduction dimensionnelle

L'objectif de cette étape est d'identifier les redondances entre nos 15 indicateurs sélectionnés. Si deux variables sont trop fortement corrélées, elles apportent la même information. Les conserver toutes les deux lors du scoring reviendrait à donner un "double poids" artificiel à une seule et même réalité, ce qui biaiserait notre classement final.

## 8.1. Méthodologie : Pearson vs Spearman

Pour obtenir une vision complète des relations entre nos variables, nous calculons deux types de coefficients :

1. **Coefficient de Pearson (Corrélation Linéaire)**
    - **Définition** : Il mesure la force d'une relation linéaire entre deux variables (une ligne droite).
    - **Usage** : Idéal pour détecter les indicateurs qui progressent de manière strictement proportionnelle.
    - **Limite** : Il est très sensible aux valeurs aberrantes (outliers), fréquentes dans les données mondiales (ex: disparité entre le PIB de la Chine et celui de l'Estonie).
2. **Coefficient de Spearman (Corrélation Monotone)**
    - **Définition** : Il se base sur le rang des valeurs (le classement) plutôt que sur les valeurs brutes.
    - **Usage** : Plus robuste que Pearson face aux outliers. Il détecte si deux variables évoluent dans le même sens, même si la relation n'est pas une ligne droite parfaite.
    - **Intérêt pour Academy** : Puisque nous comparons des pays aux échelles démographiques et économiques radicalement différentes, Spearman est souvent plus représentatif de la réalité du terrain.

## 8.2. Stratégie d'arbitrage (Pruning Strategy)

1. **Méthodologie de Fusion** : Le calcul repose sur la logique suivante pour chaque paire d'indicateurs :
    - **Calcul des valeurs absolues** : Nous transformons les coefficients de Pearson ($r$) et Spearman ($\rho$) en valeurs absolues pour nous concentrer sur la force de la liaison, quelle que soit sa direction.
    - **Sélection du maximum** : Pour chaque cellule de la matrice, nous conservons la valeur maximale : $max(|r|, |\rho|)$.
    - **Objectif** : Cette approche "conservatrice" permet de détecter une redondance dès qu'elle est soit linéaire, soit monotone, garantissant qu'aucun biais de colinéarité ne subsiste.
2. **Seuil de tolérance** : Tout couple d'indicateurs présentant un coefficient consolidé supérieur à 0.70 est considéré comme redondant. Ce seuil critique déclenche une procédure d'élimination.
3. **Critères de sélection métier** : En cas de conflit (corrélation $> 0.70$), l'arbitrage est effectué selon les priorités suivantes :
    - **Disponibilité** : Priorité à l'indicateur ayant le meilleur taux de remplissage
    - **Actionnabilité** : Priorité à l'indicateur le plus pertinent pour le modèle d'Academy (ex: favoriser les taux de complétion sur les durées théoriques).
    - **Indépendance** : Priorité à l'indicateur qui présente la corrélation moyenne la plus faible avec l'ensemble des autres variables restantes.
4. **Visualisation** : Nous utilisons une Heatmap unique basée sur cette matrice consolidée. Elle permet d'identifier d'un seul coup d'œil les clusters de variables interdépendantes, simplifiant ainsi la phase finale de nettoyage.

**Point de vigilance** : Le calcul de corrélation ignore automatiquement les pays ayant des valeurs manquantes (NaN) pour le couple de variables testé. Une gestion rigoureuse de la complétude en amont (Phase 6 & 7) est donc cruciale pour ne pas biaiser les coefficients.

## 8.3. Arbitrage et sélection finale
Une fois les corrélations identifiées par la heatmap, nous appliquons une règle de décision binaire pour chaque couple de variables dépassant le seuil de 0.70. L'objectif est de transformer une redondance statistique en un choix stratégique.

1. **Critères de suppression**
Un indicateur est supprimé s'il remplit l'une des conditions suivantes :
    - **Redondance de volume** : Si deux indicateurs mesurent la même masse démographique (ex: inscrits au collège vs inscrits au lycée), nous supprimons le moins spécifique pour éviter de surestimer le poids de la population.
    - **Redondance de structure** : Si une donnée théorique (ex: durée des études) est trop corrélée à une donnée réelle (ex: taux de réussite), la donnée théorique est écartée car elle n'apporte aucune nuance sur la performance du pays.
    - **Faible variance** : Un indicateur qui présente la même valeur pour la quasi-totalité des pays est supprimé, car il ne permet pas de classer les zones géographiques entre elles.

2. **Critères de conservation**
À l'inverse, nous choisissons de conserver un indicateur, même en cas de forte corrélation, s'il est considéré comme un "Pivot Métier" :
    - **Indicateur de faisabilité** : Les variables validant la possibilité technique de l'offre (ex: accès Internet) sont maintenues car elles représentent un risque binaire (Go/No-Go).
    - **Proxy de segmentation** : Nous conservons les indicateurs qui capturent une dimension économique ou comportementale unique (ex: part du privé, accès au supérieur), même s'ils sont statistiquement liés au niveau d'éducation global.

### 8.1.1. Méthode Pearson :

> **Note Technique : Analyse de Corrélation avec Pandas**
> L'utilisation de la méthode `.corr()` est le standard en Data Science pour quantifier la relation statistique entre deux variables. Voici le détail de l'implémentation choisie :
> 1. Paramètre de la methode `corr()` :
>   - `methode=`
>       - `'pearson'` : Calcule le coefficient de corrélation de Pearson. Il mesure la relation linéaire entre deux variables. Sa valeur varie de -1 à 1.
>       - `'spearman'` : Calcule la corrélation de rang de Spearman. Contrairement à Pearson, il est non-paramétrique et mesure la relation monotone (si une variable augmente, l'autre augmente aussi, même si ce n'est pas de façon linéaire). Il est beaucoup plus robuste face aux valeurs aberrantes
>   - `numeric_only=True` : Ce paramètre est une bonne pratique de développement. Il force Pandas à ignorer toutes les colonnes non numériques (comme les noms de pays ou codes ISO). Cela évite de lever des erreurs (TypeErrors) et optimise le temps de calcul en ne traitant que les types float64 ou int64

### 8.1.2. Méthode Spearman

```python
corr_spearman = final_academy_dataset.corr(method='spearman', numeric_only=True)
with pd.option_context(
    'display.max_columns', None,
    'display.max_rows', None
):
    display(corr_spearman)
```

### 8.2.1 Méthodologie de Fusion

> **Note Technique : Matrice de Corrélation Consolidée (Fusion Metrics)** :
>
> Cette cellule implémente une stratégie de sélection de variables conservatrice (Conservative Feature Selection). L'objectif est de s'assurer qu'aucune redondance, qu'elle soit linéaire ou non, n'échappe à l'analyse avant la phase de scoring final.
> 1. **Logique de l'Algorithme** : La fonction get_conservative_correlation_matrix repose sur trois piliers techniques :
>   - **Normalisation par Valeur Absolue (`.abs()`)** : En statistique, une corrélation de -0.90 est aussi forte qu'une corrélation de 0.90 (elles indiquent toutes deux une dépendance totale). Nous transformons les résultats pour ne manipuler que la force de la relation, simplifiant ainsi l'interprétation visuelle.
>   - **Maximisation Élémentaire (`np.maximum`)** : Contrairement à une moyenne, le choix du maximum est une approche de "pire scénario". Si Pearson détecte 0.50 mais que Spearman détecte 0.85, la fonction retiendra 0.85. Cela permet de capturer des relations monotones non-linéaires que Pearson sous-estimerait.
>   - **Réduction de la Charge Cognitive** : En fusionnant deux matrices de 225 cellules en une seule, nous divisons par deux l'effort d'audit visuel tout en augmentant la fiabilité de la détection.
> 2. **Visualisation par Heatmap** :
>   - **Palette `Reds`** : Contrairement à la palette divergente (bleu-rouge), nous utilisons ici une palette séquentielle (dégradés de rouge). Puisque toutes nos valeurs sont positives (valeurs absolues), l'intensité du rouge indique directement le degré de redondance.
>   - Paramètre `fmt=".2f`" : Ce formatage assure une lecture claire des coefficients avec deux décimales, évitant ainsi l'encombrement visuel de la matrice.

```python
def get_conservative_correlation_matrix(df):

    pearson_abs = df.corr(method='pearson', numeric_only=True).abs()
    spearman_abs = df.corr(method='spearman', numeric_only=True).abs()

    combined_corr = np.maximum(pearson_abs, spearman_abs)

    return combined_corr

combined_matrix = get_conservative_correlation_matrix(final_academy_dataset)

plt.figure(figsize=(16, 10))
plt.title("Matrice de Corrélation Consolidée (Max de |Pearson| et |Spearman|)")
sns.heatmap(combined_matrix, annot=True, cmap='Reds', fmt=".2f")
plt.show()
```

> **Note Technique : Décorateur Visuel Conditionnel**
>   - **Logic de rendu** : Plutôt que de modifier la fonction heatmap elle-même (qui ne supporte pas nativement les styles par cellule), on utilise l'objet ax (l'axe Matplotlib) comme un canvas sur lequel on vient dessiner par-dessus.
>   - **Système de coordonnées** : Dans une Heatmap Seaborn, chaque case d'index (i, j) correspond à une unité de dessin 1x1 sur l'axe. C'est pourquoi on utilise (j, i) pour les coordonnées du rectangle (Matplotlib utilise l'ordre x, y).
>   - **`lw` (Line Width)** : On règle l'épaisseur à 4 pour que le contour soit bien visible même avec les annotations numériques à l'intérieur.

```python
import matplotlib.patches as patches

plt.figure(figsize=(16, 10))
plt.title("Matrice Consolidée avec Alertes de Redondance (|r| > 0.70)")
ax = sns.heatmap(combined_matrix, annot=True, cmap='Reds', fmt=".2f")
for i in range(len(combined_matrix)):
    for j in range(len(combined_matrix.columns)):
        if combined_matrix.iloc[i, j] >= 0.7 and i != j:
            ax.add_patch(patches.Rectangle((j, i), 1, 1, fill=False, edgecolor='blue', lw=4))

plt.show()
```

# 9. Méthodologie d'Analyse Descriptive et Quantitative

Cette section détaille la démarche analytique utilisée pour évaluer le potentiel de chaque pays. L'objectif est de passer d'une compréhension statistique individuelle de chaque indicateur à une synthèse globale permettant de classer les marchés prioritaires pour Academy.

## 9.1. Analyse Descriptive des Indicateurs
Conformément aux instructions, nous avons automatisé l'étude de chaque variable pour garantir une analyse homogène et rigoureuse.
1. **Statistiques de centralité et de dispersion** :
    - **La Moyenne (Mean)** : Valeur moyenne mondiale, sensible aux valeurs extrêmes.
    - **La Médiane (50%)** : Valeur centrale séparant l'échantillon en deux. La comparaison entre moyenne et médiane nous permet de détecter instantanément l'asymétrie (Skewness) de la distribution.
    - **Les Quantiles (25% et 75%)** : Ils définissent les seuils de performance. Le 3ème quartile (75%) est particulièrement crucial car il marque l'entrée dans le "Top 25%" des pays les plus performants pour un critère donné.
2. **Visualisation des Distributions** :
    - **La Densité (KDE)** : Une courbe lissée qui révèle la "forme" mathématique de la distribution (normale, bimodale ou étirée).
    - **La Concentration** : L'identification des zones où se regroupent la majorité des pays.
    - **Les Outliers** : Détection visuelle des pays atypiques qui présentent un potentiel hors norme (ex: géants démographiques).

## 9.2. Stratégie de Sélection Quantitative

Pour répondre à la demande de Mark et identifier les pays cibles, nous ne nous contentons pas d'un seul critère. Nous proposons une méthode basée sur la Récurrence dans les Tops de Performance.

1. **Le principe de l'algorithme :**
    - **Extraction des Leaders** : Pour chaque indicateur, nous isolons les pays situés dans le haut du classement (Top 15).
    - **Calcul de la Récurrence** : Nous comptabilisons le nombre de fois qu'un pays apparaît dans ces listes d'excellence. Un pays apparaissant dans 8 tops sur 10 est statistiquement plus robuste qu'un pays premier sur un seul critère mais absent des autres.
    - **Filtrage par Faisabilité** : Nous appliquons un filtre strict sur les indicateurs "bloquants" (ex: accès Internet). Un pays ne peut être retenu que s'il dépasse un seuil de connectivité minimale, quel que soit son volume d'élèves.
    - **Avantages de cette méthode :**
        - **Neutralité** : Elle limite l'impact d'un seul indicateur qui pourrait être surévalué.
        - **Stabilité** : Elle favorise les pays présentant un profil équilibré (bonne infrastructure + bon volume + bonne éducation).
        - **Actionnabilité** : Elle fournit une liste finale "clé en main" classée par score de confiance.

## 9.3. Normalisation des données (Min-Max Scaling)

Pour transformer nos analyses en une recommandation concrète, nous utilisons un algorithme de scoring pondéré. Cette méthode est plus précise qu'un simple classement par indicateur car elle permet de hiérarchiser l'importance de chaque variable selon le business model d'Academy.

1. **Normalisation des données**
Nos indicateurs utilisent des unités hétérogènes (millions d'élèves, pourcentages, années). Pour les comparer équitablement, nous appliquons une Normalisation Min-Max.
   - **Principe** : Chaque valeur est ramenée sur une échelle de 0 à 1.
   - **Résultat** : Le pays le moins performant sur un critère reçoit la note de 0, et le meilleur la note de 1. Cette étape est indispensable pour que les gros volumes (ex: population) n'écrasent pas les ratios (ex: % d'accès Internet).

2. **Attribution des Coefficients** :
Nous avons défini des poids stratégiques pour refléter les priorités d'investissement d'Academy
3. **Calcul du Score Final** :
Le score final d'un pays est la somme de ses notes normalisées multipliées par leurs coefficients respectifs.

$$Score_{pays} = \sum (ValeurNormalisée_{i} \times Poids_{i})$$

Cette méthode garantit que les pays recommandés présentent le meilleur équilibre entre une infrastructure solide et un volume de clientèle important.

### 9.1.1. Statistiques de centralité et de dispersion

> **Note technique : Analyse descriptive automatisée**
> Cette fonction constitue la brique de base du Data Profiling de l'étape 9. Elle permet de passer d'une vision tabulaire à une compréhension statistique de la distribution de chaque indicateur.
>
> 1. Logique Statistique
>   - `describe()` : Utilise l'implémentation Pandas pour extraire les moments d'ordre 1 (moyenne) et les mesures de dispersion (écart-type, quartiles). Les quartiles $Q_1$ (25%) et $Q_3$ (75%) sont essentiels pour identifier les pays "performants" (au-dessus du 75ème percentile).
>   - Centralité vs Asymétrie : Le traçage simultané de la Moyenne (Mean) et de la Médiane (50%) permet de détecter visuellement le biais :
>       - Si `Mean` > `Median` : Distribution étirée vers la droite (présence d'outliers à fortes valeurs).
>       - Si `Mean` ≈ `Median` : Distribution symétrique.
> 2. **Implémentation Graphique**
>   - `sns.histplot` : Choisi pour sa capacité à combiner l'histogramme (fréquences discrètes) et le KDE (Kernel Density Estimate). Le KDE lisse la distribution pour révéler la loi de probabilité sous-jacente.
>   - **Gestion des données manquantes** : L'utilisation de .dropna() est critique ici pour éviter les erreurs de calcul du backend de Seaborn lors de la génération de la courbe de densité.
>   - `axvline` : L'ajout de lignes verticales (Vertical Axes Lines) transforme un graphique passif en un outil d'aide à la décision, facilitant la lecture des seuils critiques.
> 3. **Paramétrage de l'affichage**
>   - `figsize=(12, 5)` : Définit un ratio d'aspect large, optimal pour observer l'étalement des données sur l'axe X, particulièrement pour les indicateurs avec une "longue traîne" comme les volumes d'inscriptions.
>   - `plt.show()` : Force le vidage du buffer graphique à chaque appel, garantissant que chaque indicateur dispose de son propre espace de rendu sans superposition.

```python
def analyze_and_plot_indicator(df, indicator_name):
    stats = df[indicator_name].describe()
    display(stats.to_frame().T)
    plt.figure(figsize=(12, 5))
    sns.histplot(df[indicator_name].dropna(), kde=True, color="teal")
    plt.axvline(stats['mean'], color='red', linestyle='--', label=f"Moyenne: {stats['mean']:.2f}")
    plt.axvline(stats['50%'], color='orange', linestyle='-', label=f"Médiane (Q2): {stats['50%']:.2f}")

    plt.title(f"Distribution et Dispersion : {indicator_name}")
    plt.legend()
    plt.show()
```

> **Note Technique : Guide d'interprétation des Distributions (KDE & Histogrammes)**
> Cette note détaille la méthodologie de lecture des graphiques générés par la fonction `analyze_and_plot_indicator`. L'objectif est de traduire des formes géométriques en opportunités business pour Academy.
> 1. **Analyse de la Forme**
>   - **Asymétrie à droite** : La "bosse" est à gauche, avec une longue traîne vers les valeurs élevées.
>       - Signification : La majorité des pays ont des scores faibles. Les pays dans la traîne sont des outliers à haut potentiel (ex: volumes d'étudiants massifs).
>   - **Asymétrie à gauche** : La "bosse" est à droite.
>       - Signification : La majorité des pays ont déjà atteint des scores élevés. C'est le signe d'un marché mature ou d'un pré-requis largement rempli (ex: Taux d'alphabétisation).
>   - **Distribution Bimodale** : Présence de deux sommets.
>       - Signification : Le monde est divisé en deux groupes distincts (ex: pays développés vs pays en développement), sans véritable "milieu".
> 2. **Lecture des Indicateurs de Centralité**

### 9.2. Stratégie de Sélection Quantitative

> **Note Technique : Algorithme de Recommandation par Récurrence**
>
> Cette fonction constitue le moteur de décision final pour répondre à la problématique d'Academy. Elle transforme les performances individuelles par indicateur en un score de fiabilité globale.
>
> 1. **Logique Algorithmique**
>   - **Isolement des élites** : Pour chaque dimension (colonne numérique), la fonction effectue un tri descendant (sort_values) et extrait les $n$ meilleurs pays ($n=15$ par défaut).
>   - **Comptage des occurrences** : Un dictionnaire de fréquences (occurrence_count) incrémente le score d'un pays chaque fois qu'il apparaît dans un "Top 15".
>   - **Agrégation** : Le score final représente le nombre de critères pour lesquels le pays est considéré comme un leader mondial.
> 2. **Détails de l'implémentation**
>   - **Sélectivité dynamique** : Utilisation de `select_dtypes(include=['number'])` pour ignorer automatiquement les métadonnées (`Country Name`, `Code`) et ne traiter que les variables de scoring.
>   - **Structure de données** : Le passage par un dictionnaire pour le comptage assure une complexité temporelle de $O(n \times m)$ (où $n$ est le nombre de colonnes et $m$ la taille du Top), ce qui est très performant sur des datasets de cette taille.
>   - Sortie : La conversion finale en `DataFrame` permet une exploitation immédiate (export CSV, visualisation ou filtrage complémentaire).

```python
def get_top_countries_recommendation(df, top_n=15):
    occurrence_count = {}
    numeric_cols = df.select_dtypes(include=['number']).columns

    for col in numeric_cols:
        top_list = df.sort_values(by=col, ascending=False).head(top_n)['Country Name'].tolist()

        for country in top_list:
            occurrence_count[country] = occurrence_count.get(country, 0) + 1
    recommendation_df = pd.DataFrame(
        list(occurrence_count.items()),
        columns=['Country Name', 'Score_Recurrence']
    ).sort_values(by='Score_Recurrence', ascending=False)

    return recommendation_df

final_recommendation = get_top_countries_recommendation(final_scoring_dataset)
display(final_recommendation.head(15))
```

> **Note Technique : Justification des Coefficients de Pondération**
>
> 1. Les multiplicateurs critiques ($w \ge 2.5$)
>   - Internet users : Sans connectivité, le business model s'effondre. C'est l'indicateur "pivot".
>   - Enrolment (Secondary) : C'est l'indicateur de volume pur. Un pays avec un excellent taux de pénétration mais seulement 10 000 élèves au total sera moins rentable qu'un marché de masse.
> 2. Les leviers de conversion ($w \in [1.0, 2.0]$)
>   - Tertiary & Barro-Lee : Ces indicateurs mesurent l'appétence des familles pour les études longues. Plus le taux d'accès au supérieur est élevé, plus le besoin de soutien scolaire pour y parvenir est fort.
>   - Private institutions & Literacy : Ils valident la solvabilité et le niveau d'éducation des parents (décideurs d'achat).
> 3. Les variables de contexte ($w = 0.5$)
>   - Vocational / Duration : Bien qu'utiles pour la logistique des cours, ces variables n'indiquent pas directement le potentiel de chiffre d'affaires. Un poids faible évite qu'elles ne parasitent le classement final.

```python
weights = {
    # PRIORITÉ 1 : Infrastructure & Faisabilité (Indispensable pour l'e-learning)
    'Internet users (per 100 people)': 3.0,

    # PRIORITÉ 2 : Volume du marché cible (Taille du réservoir de clients)
    'Enrolment in secondary education, both sexes (number)': 2.5,

    # PRIORITÉ 3 : Potentiel de croissance & CSP+ (Capacité à investir dans le supérieur)
    'Gross enrolment ratio, tertiary, both sexes (%)': 2.0,
    'Barro-Lee: Percentage of population age 15+ with secondary schooling. Completed Secondary': 1.5,

    # PRIORITÉ 4 : Qualité & Écosystème (Fiabilité du flux d'élèves)
    'Lower secondary completion rate, both sexes (%)': 1.0,
    'Literacy rate, population 25-64 years, both sexes (%)': 1.0,
    'Percentage of enrolment in secondary education in private institutions (%)': 1.0,

    # PRIORITÉ 5 : Indicateurs de structure (Contexte moins impactant sur le CA)
    'School life expectancy, primary and lower secondary, both sexes (years)': 0.5,
    'Percentage of students in secondary education enrolled in vocational programmes, both sexes (%)': 0.5,
    'Theoretical duration of secondary education (years)': 0.5
}
```

> **Note Technique : Optimisation et Intégration**
>
> Algorithme de Scoring Pondéré : Pour classer les pays, nous utilisons un modèle d'agrégation linéaire basé sur des poids métier. Cette approche permet de hiérarchiser les critères selon les priorités stratégiques d'Academy.
> **Etape de calcul** :
>   - **Normalisation Min-Max** : Aligne tous les indicateurs sur une échelle de $[0, 1]$ pour neutraliser les différences d'unités (pourcentages vs millions d'habitants).
>   - **Application des coefficients** : Multiplication de chaque valeur normalisée par son poids d'importance stratégique (ex: Internet x3).
>   - **Calcul du Score Final** : Somme pondérée de l'ensemble des indicateurs pour obtenir une note globale d'attractivité.

```python
from sklearn.preprocessing import MinMaxScaler

def calculate_final_score(df, weights):
    numeric_cols = list(weights.keys())
    scaler = MinMaxScaler()
    df_norm = df.copy()
    df_norm[numeric_cols] = scaler.fit_transform(df[numeric_cols])

    df['Final_Score'] = sum(df_norm[col] * weights[col] for col in numeric_cols)

    return df.sort_values(by='Final_Score', ascending=False)
```

```python
top_countries = calculate_final_score(final_scoring_dataset, weights)
display(top_countries[['Country Name', 'Final_Score']].head(15))
```
